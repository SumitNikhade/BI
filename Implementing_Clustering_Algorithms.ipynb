```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate synthetic dataset with 3 clusters
n_samples = 300
n_features = 2
n_clusters = 3
random_state = 42

X, _ = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=random_state)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=random_state, n_init=10)
y_kmeans = kmeans.fit_predict(X)

# Plot the clustered data
plt.figure(figsize=(8, 6))
for i in range(n_clusters):
    plt.scatter(X[y_kmeans == i, 0], X[y_kmeans == i, 1], label=f'Cluster {i+1}')

# Plot cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroids')

plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

# Initialize the Agglomerative Clustering model
agg_clustering = AgglomerativeClustering(n_clusters=4)

# Fit the model to the data and get cluster assignments
y_agg = agg_clustering.fit_predict(X)

# Visualize the clusters
plt.figure() # Create a new figure for the new plot
plt.scatter(X[:, 0], X[:, 1], c=y_agg, s=50, cmap='viridis')

plt.title('Agglomerative Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
